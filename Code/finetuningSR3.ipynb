```python
import json
import re
import os

config_path = '/kaggle/working/Image-Super-Resolution-via-Iterative-Refinement/config/sr_sr3_64_512.json'

def load_config(path):
    with open(path, 'r') as f:
        content = f.read()
    content = re.sub(r'//.*', '', content)
    return json.loads(content)

config = load_config(config_path)

config['datasets']['train']['dataroot'] = 'dataset/celebahq_64_512'
config['datasets']['val']['dataroot'] = 'dataset/celebahq_64_512'
config['path']['resume_state'] = "I832500_E52"

config['train']['n_iter'] = 850000
config['datasets']['train']['batch_size'] = 4
config['train']['val_freq'] = 1000
config['train']['save_checkpoint_freq'] = 1500

with open(config_path, 'w') as f:
    json.dump(config, f, indent=4)
```
